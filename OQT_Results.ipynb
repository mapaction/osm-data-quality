```python
import geopandas as gpd
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
f = gpd.read_file(f'results/cameroon_ad1_mapoc.geojson')

column_data = 'report.result.value'
column_name = 'report.metadata.name'

def create_plot(column_data, column_name): 
    title = f[column_name][0]
    fig, ax = plt.subplots(1, 1)

    f.plot(column=column_data,
            ax=ax,
            cmap='Greens',
            legend=True,
            vmax=1,
            vmin=0,
            legend_kwds={'label': "Confidence Index Value"})

    ax.axis('off')
    plt.title(title)
    plt.show()

create_plot(column_data, column_name)
```

```python
for i in range(1,10):
    column_data = f'indicators.{i}.result.value'
    column_name = f'indicators.{i}.metadata.name'
    create_plot(column_data, column_name)
```
